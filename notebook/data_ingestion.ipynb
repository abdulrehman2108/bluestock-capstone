```python
import pandas as pd
import os
```

```python
data_path = "C:/Users/abdul/OneDrive/Desktop/Bluestock_capstone/data/raw"

csv_files = [file for file in os.listdir(data_path) if file.endswith(".csv")]

datasets= {}

print("Data ingestion started...")

for file in csv_files:
    file_path = os.path.join(data_path, file)
    
    
    try:

        print("Loading dataset")

        df=pd.read_csv(file_path)
        datasets[file]=df
        print(f"\nLoaded: {file}")


        print("Shape:")
        print(df.shape)

        print("Data Types")
        print(df.dtypes)

        print("Head")
        print(df.head())

        print("Missing Values")
        print(df.isnull().sum())

        print("duplicated")
        print(df.duplicated().sum())

        print("Summary Statistics")
        print(df.describe())

        print("\n")

    except Exception as e:
        
        print(f"Error loading {file}: {e}")
        print("\n")

print(f"Loaded {len(datasets)} datasets successfully!")
```

```python
if ("fund_master.csv" in datasets):
    fund_master = datasets["fund_master.csv"]
    
    fund_master["fund_house"] = (
        fund_master["scheme_name"].str.split().str[0]
    )

    print("Unique fund houses")
    print(fund_master["fund_house"].unique())
```

```python
nav_files = [file for file in csv_files if "_nav" in file]

nav_history = pd.DataFrame()

for file in nav_files:

    df = datasets[file]

    nav_history = pd.concat([nav_history, df])
if "Fund_Master.csv" in datasets:

    master_codes = set(fund_master["schemeCode"])

    nav_codes = set(nav_history["scheme_code"])

    missing_codes = master_codes - nav_codes

    print("\nMissing Scheme Codes:")
    print(len(missing_codes))
```
