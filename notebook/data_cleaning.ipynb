```python
perf_df=pd.read_csv(
    "C:/Users/abdul/OneDrive/Desktop/Bluestock_capstone/data/processed/scheme_performance.csv")

perf_df.head()
perf_df.info()
```

```python
perf_df.isnull().sum()
```

```python
perf_df.columns
```

```python
numeric_cols = [
    "return_1yr_pct",
    "return_3yr_pct",
    "return_5yr_pct",
    "benchmark_3yr_pct",
    "alpha",
    "beta",
    "sharpe_ratio",
    "sortino_ratio",
    "std_dev_ann_pct",
    "max_drawdown_pct",
    "aum_crore",
    "expense_ratio_pct",
    "morningstar_rating"
]

for col in numeric_cols:

    perf_df[col] = pd.to_numeric(
        perf_df[col],
        errors="coerce"
    )
```

```python
return_cols = [
    "return_1yr_pct",
    "return_3yr_pct",
    "return_5yr_pct"
]

for col in return_cols:

    anomalies = perf_df[
        (perf_df[col] < -100) |
        (perf_df[col] > 200)
    ]

    print(f"\nAnomalies in {col}")
    print(anomalies[[col]])
```

```python
invalid_expense = perf_df[
    (perf_df["expense_ratio_pct"] < 0.1) |
    (perf_df["expense_ratio_pct"] > 2.5)
]

print(invalid_expense)
```

```python
invalid_rating = perf_df[
    ~perf_df["morningstar_rating"].isin(
        [1, 2, 3, 4, 5]
    )
]

print(invalid_rating)
```

```python
invalid_beta = perf_df[
    (perf_df["beta"] < 0) |
    (perf_df["beta"] > 5)
]

print(invalid_beta)
```

```python
perf_df["fund_house"] = (
    perf_df["fund_house"]
    .str.strip()
    .str.title()
)
```

```python
perf_df["category"] = (
    perf_df["category"]
    .str.strip()
    .str.title()
)
```

```python
perf_df["risk_grade"] = (
    perf_df["risk_grade"]
    .str.strip()
    .str.title()
)
```

```python
perf_df = perf_df.drop_duplicates(
    subset=["amfi_code"]
)
```
